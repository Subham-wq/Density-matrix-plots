```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the real and imaginary parts of the density matrix
real_part = np.array([[0.7165, -0.0545], [-0.0545, 0.2835]])
imag_part = np.array([[0, -0.337], [0.337, 0]])

# Define the coordinates
x = np.array([0, 1, 0, 1])
y = np.array([0, 0, 1, 1])

# Create a function to plot the matrix
def plot_matrix(matrix, title, ax):
    z = matrix.flatten()
    ax.bar3d(x, y, np.zeros_like(z), 0.5, 0.5, z, shade=True)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Value')
    ax.set_title(title)
    # Set the limits to ensure the plot looks well-contained
    ax.set_xlim([0, 1.5])
    ax.set_ylim([0, 1.5])
    ax.set_zlim([min(z) - 0.1, max(z) + 0.1])

# Create the figure
fig = plt.figure(figsize=(12, 6))

# Plot the real part
ax1 = fig.add_subplot(121, projection='3d')
plot_matrix(real_part, 'Real Part of Experimental density matrix ', ax1)

ax1.view_init(elev=20, azim=30)
# Plot the imaginary part
ax2 = fig.add_subplot(122, projection='3d')
plot_matrix(imag_part, 'Imaginary Part of Experimental density matrix', ax2)

# Rotate the imaginary part plot
ax2.view_init(elev=20, azim=30)  # Adjust elev and azim to rotate the view

plt.show()

```
